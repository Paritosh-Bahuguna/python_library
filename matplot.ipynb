```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python

plt.plot(x,y)

plt.xlabel('angle')
plt.ylabel('value')
plt.title('sine wave')

plt.show()
```

```python
# parabola
```

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g*')
```

```python
x=np.linspace(-5,5,200)
plt.plot(x,np.sin(x),'g*')
plt.plot(x,np.cos(x),'b')
plt.show()
```

```python
#bar plot
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
lang=['English','Hindi','German','Japanese','Nepali']
people=[100,90,60,30,11]
ax.bar(lang,people)
plt.xlabel('language')
plt.ylabel('people ')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
lang=['English','Hindi','German','Japanese','Nepali']
people=[100,90,60,30,11]
ax.pie(people,labels=lang,autopct='%1.1f%%')
plt.show()
```

```python
#Scatter plot
```

```python
x=np.linspace(0,20,100)
y=np.sin(x)
z=np.cos(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```
